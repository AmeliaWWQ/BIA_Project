```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
def crop_img(img):

    # Threshold image
    #img_binary = (img[:, :, 1] > threshold).astype(np.uint8)*255
    img_binary = (img[:, :, 1] > 2).astype(np.uint8)*255
    
    # Gather image dimensions
    r1, c1 = img_binary.shape

    # LE = Left Edge
    # RE = Right Edge
    # UE = Superior Edge
    # IE = Inferior Edge

    LE = 0
    RE = 0
    UE = 0
    IE = 0

    # Find upper edge cropping limit (UE)
    UE = 0
    for yn in range(r1):    
        if 255 in img_binary[yn, :]:
            UE = yn
            break

    # Find lower edge cropping limit (IE)
    IE = r1
    for yn in range(r1 - 1, -1, -1):    
        if 255 in img_binary[yn, :]:
            IE = yn + 1
            break

    # Find left edge cropping limit (LE)
    LE = 0
    for xn in range(c1):
        if 255 in img_binary[:, xn]:
            LE = xn
            break

    # Find right edge cropping limit (RE)
    RE = c1
    for xn in range(c1 - 1, -1, -1):
        if 255 in img_binary[:, xn]:
            RE = xn + 1
            break

    # Crop the image
    cropped_image_binary = img_binary[UE:IE, LE:RE]
    cropped_image = img[UE:IE, LE:RE]

    return cropped_image
```

```python
Crop_image_set = ['TRAIN086667.JPG', 'TRAIN086707.JPG', 'TRAIN086721.JPG', 'TRAIN086732.JPG', 'TRAIN086773.JPG', 
                  'TRAIN086778.JPG', 'TRAIN086783.JPG', 'TRAIN086784.JPG', 'TRAIN086804.JPG']

for img_filename in Crop_image_set:
    img = imageio.imread(img_filename)
    plt.imshow(img)
    plt.show()
    cropped_img = crop_img(img)
    plt.imshow(cropped_img)
    plt.show()
```
